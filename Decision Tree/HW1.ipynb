```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import preprocessing,metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
leCrime = preprocessing.LabelEncoder()
category = leCrime.fit_transform(train.Category)#標籤
train = train[['Dates', 'DayOfWeek', 'PdDistrict', 'X', 'Y' ]]
train.isnull().sum()#看有無空值
```

```python
print(train)
```

```python
#因子化特徵
days = pd.get_dummies(train.DayOfWeek)
district = pd.get_dummies(train.PdDistrict)
hour = pd.to_datetime(train['Dates']).dt.hour
hour = pd.get_dummies(hour)
#hour = train.Dates.dt.hour
#hour = pd.get_dummies(hour) 
#组合特征
train = pd.concat([hour, days, district], axis=1)
train['category']=category

days = pd.get_dummies(test.DayOfWeek)
district = pd.get_dummies(test.PdDistrict)
hour = pd.to_datetime(test['Dates']).dt.hour
hour = pd.get_dummies(hour)
test = pd.concat([hour, days, district], axis=1)
#testData = pd.concat([hour, days, district], axis=1)
train
```

```python
features = ['Friday', 'Monday', 'Saturday', 'Sunday', 'Thursday', 'Tuesday',
'Wednesday', 'BAYVIEW', 'CENTRAL', 'INGLESIDE', 'MISSION',
'NORTHERN', 'PARK', 'RICHMOND', 'SOUTHERN', 'TARAVAL', 'TENDERLOIN']
hourFea = [x for x in range(0,24)]
features = features + hourFea
X = train.drop(['category'],axis=1 )
Y = train['category']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size=0.75,test_size=0.25)
model = DecisionTreeClassifier(max_depth=3)#建立模型
model = model.fit(x_train, y_train)#訓練模型
y_pred = model.predict(x_test)#預測測試集
```

```python
print(classification_report(y_test, y_pred))
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```
