```python
from sklearn import datasets, metrics
import pandas as pd
import numpy as np
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
en_stops = set(stopwords.words('english'))
```

```python
f = open('training_label.txt',encoding="utf-8")
f2 = open('testing_label.txt',encoding="utf-8")
num = 10000
train = []
train_label = []
test = []
test_label = []
l = 0
for line in f:
    if(l<num):#取前10000筆
        train.append(line[10:-1])#文字部分
        train_label.append(line[0])#label
        l = l+1
        
    else:break

for line in f2:
    if(line=="\n"):#如果讀到都是空的繼續往下讀
        continue
    test.append(line[6:-1])#文字部分
    test_label.append(line[0])#label
```

```python
df_train_data = pd.DataFrame(train)#dataframe
df_train_label = pd.DataFrame(train_label)#dataframe
df_train = pd.concat([df_train_label,df_train_data], axis=1)#合併label和文字
df_train.columns = ['label','data']#設定column名稱
df_test_data = pd.DataFrame(test)#dataframe
df_test_label = pd.DataFrame(test_label)#dataframe
df_test = pd.concat([df_test_label,df_test_data], axis=1)
df_test.columns = ['label','data']#設定column名稱
```

df_train

```python
df_test
```

```python
train_a = []
train_b = [[] for i in range(num)]
test_a = []
test_b = [[] for i in range(df_test.shape[0])]
for i in range(num):
    train_a.append(df_train['data'][i].split())#將一句話分割每個文字
    
for i in range(df_test.shape[0]):
    test_a.append(df_test['data'][i].split())#將一句話分割每個文字 
test_a
```

```python
for i in range(num):
    for word in train_a[i]:
        if word not in en_stops:#如果讀到的字不存在於停用詞則存起來
            train_b[i].append(word)
            
for i in range(num):
    train_b[i] = " ".join(train_b[i])#合併
    
for i in range(df_test.shape[0]):
    for word in test_a[i]:
        if word not in en_stops:#如果讀到的字不存在於停用詞則存起來
            test_b[i].append(word)
            
for i in range(df_test.shape[0]):
    test_b[i] = " ".join(test_b[i])#合併
test_b
```

```python
from string import digits #去除數字
 
for i in range(num):
    remove_digits = str.maketrans('', '', digits)
    train_b[i] = train_b[i].translate(remove_digits)
train_b

for i in range(df_test.shape[0]):
    remove_digits = str.maketrans('', '', digits)
    test_b[i] = test_b[i].translate(remove_digits)
# train_b
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation,Flatten
from keras.layers.embeddings import Embedding
from keras.layers.recurrent import LSTM

model = Sequential()
model.add(Embedding(output_dim=32,
                    input_dim=3800, 
                    input_length=13047))
model.add(Dropout(0.7))

model.add(LSTM(32))
model.add(Dense(units=256,
                activation='relu' ))
model.add(Dropout(0.7))
model.add(Dense(units=1,
                activation='sigmoid' ))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])

train_history =model.fit(tfidf,label_all ,batch_size=100, 
                         epochs=10,verbose=2,
                         validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
def show_train_history(train_history,train,validation):
    plt.plot(train_history.history[train])
    plt.plot(train_history.history[validation])
    plt.title('Train History')
    plt.ylabel(train)
    plt.xlabel('Epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```

```python
show_train_history(train_history,'acc','val_acc')
```

```python
show_train_history(train_history,'loss','val_loss')
```
