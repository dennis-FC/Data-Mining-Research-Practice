```python
import pandas as pd
from datetime import datetime
import numpy as np
data = pd.read_csv("DATA.csv")#讀資料
data = data.drop(["測站"],1)#不要測站這一欄
data = pd.DataFrame(data)

month = data["日期"].apply(lambda x:datetime.strptime(x,'%Y/%m/%d'))
month = month.map(lambda x: x.month)
data["月"] = month#新增月這一欄存放月
#除了十，十一，十二月其他都不要
indexNames = data[ (data['月'] ==1) | (data['月'] ==2) | (data['月'] ==3) | (data['月'] ==4) | (data['月'] ==5) | (data['月'] ==6) | (data['月'] ==7)| (data['月'] ==8) | (data['月'] ==9)].index
data.drop(indexNames , inplace=True)
data = data.drop(["月"],1)#把月這一欄刪除
data.reset_index(inplace=True)#重設index
data
```

```python
data = data.fillna(0)
data.replace("NR",0,inplace=True)#替換NR為0
a = ""
b = 0.0
c = 0.0
for i in range(1656):
    for j in range(3,27):
        try:
            float(data.loc[i][j])>=0#如果資料正常不處理
        except:#當資料怪怪進入這裡
            a = data.loc[i][j]#怪怪的資料
            if j-1==2:#往前找如果前面沒有了，往前一天繼續找
                b = float(data.loc[i-18][26])#怪怪資料的左邊
            else:
                b = float(data.loc[i][j-1])
            jj = j
            ii = i
        
            while True:#找怪怪資料的右邊
                try:
                    jj = jj+1
                    if jj > 26:#如果右邊沒資料了
                        ii = ii+18#往下一天找
                        jj = 3
                    if float(data.loc[ii][jj])>=0:
                        c = float(data.loc[ii][jj])#找到怪怪資料右邊的正常資料
                        break

                except:
                    pass#continue

            data.replace(a,str((b+c)/2),inplace=True)#替換怪怪資料為左加右的平均
        
data
```

```python
data = data.drop(['日期'],1)
data = data.drop(['index'],1)
data = data.drop(['測項'],1)
```

```python
#分割成十月十一月並且每18個特徵合併
train = data.iloc[:18,:]
for i in range(18,1098,18):
    cut = data.iloc[i:i+18,:]#切資料
    cut.index = range(0,18)
    train = pd.concat([train,cut],axis = 1,join_axes = [train.index])#合併
id = []
for i in range(len(train.columns)):
    id.append(str(i))
train.columns = id
train
```

```python
#分割成十二月並且每18個特徵合併
test = data.iloc[1098:1116,:]
for i in range(1116,1656,18):
    cut = data.iloc[i:i+18,:]
    cut.index = range(1098,1116)
    #print(cut)
    test = pd.concat([test,cut],axis = 1,join_axes = [test.index])
id = []
for i in range(len(test.columns)):
    id.append(str(i))
test.columns = id
test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

trainPM = np.array(train)[9]#PM2.5
testPM = np.array(test)[9]#PM2.5

X_train = []#訓練集
for i in range(0,len(trainPM)-6,1):
    X_train.append(trainPM[i:i+6])

Y_train = []#訓練集的答案
for i in range(6,len(trainPM)):
    Y_train.append(trainPM[i])
    
X_test = []#測試集
for i in range(0,len(testPM)-6,1):
    X_test.append(testPM[i:i+6])
    
Y_test = []#測試集的答案
for i in range(6,len(testPM)):
    Y_test.append(float(testPM[i]))

```

```python
model = LinearRegression()#建模
model.fit(X_train,Y_train)#合併資料
pred = model.predict(X_test)#預測
MAE = mean_absolute_error(Y_test,pred)
print("只有PM2.5")
print("線性回歸:",MAE)

tree = RandomForestRegressor(300)#建模
tree.fit(X_train,Y_train)
pred = tree.predict(X_test)
MAE = mean_absolute_error(Y_test,pred)
print("隨機森林:",MAE)
```
